```python
from ultralytics import YOLO
import torch
from pathlib import Path

model = YOLO("/home/king/Local/runs/detect/runs/part_train/part_detector/weights/best.pt")
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)

def check_part(image_path, conf_threshold=0.5):
    results = model.predict(image_path, device=device, verbose=False, conf=conf_threshold)
    if not results:
        return False
    res = results[0]
    if res.boxes is None or len(res.boxes) == 0:
        return False
    for box in res.boxes:
        cls_id = int(box.cls[0]) if hasattr(box, "cls") else -1
        conf = float(box.conf[0]) if hasattr(box, "conf") else 0.0
        if cls_id == 0 and conf >= conf_threshold:
            return True
    return False

images_dir = Path("/home/king/Downloads/Training")  # same as PART_IMAGES_DIR
image_exts = {".jpg", ".jpeg", ".png", ".bmp"}
image_paths = [p for p in images_dir.rglob("*") if p.suffix.lower() in image_exts]

for img_path in sorted(image_paths):
    present = check_part(str(img_path))
    print(img_path.name, "->", "PART_PRESENT" if present else "NO_PART")
```
